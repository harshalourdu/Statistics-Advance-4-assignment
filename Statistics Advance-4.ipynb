# Q2: Differentiate between one-tailed and two-tailed tests.
One-tailed test: In a one-tailed test, the alternative hypothesis specifies a direction (either greater than or less than). It is used when you are only interested in detecting an effect in one direction.

Example: A drug manufacturer tests whether their new drug increases the recovery rate (i.e., greater than the current standard).
Two-tailed test: A two-tailed test is used when the alternative hypothesis does not specify a direction. You are testing for the possibility of an effect in both directions.

Example: A researcher is testing whether the average height of men differs from the population mean. The effect can be either an increase or decrease.



# Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.
Type 1 Error (False Positive): This occurs when you reject the null hypothesis when it is actually true.
Example: A drug trial finds that the new drug is effective when it actually is not.
Type 2 Error (False Negative): This occurs when you fail to reject the null hypothesis when it is actually false.
Example: A drug trial fails to detect the effectiveness of a new drug when it actually is effective.
